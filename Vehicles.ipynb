# Outliers

```python
sns.boxplot(x=df["price"])
plt.show()
```

```python
Q1=df["price"].quantile(.25)
Q3=df["price"].quantile(.75)
IQR=Q3-Q1
print ("Low Limit: " , (Q1-.2*IQR))
print ("High Limit: " , (Q3+3*IQR))
```

```python
IQR
```

```python
left_limit=Q1-.2*IQR
left_limit
```

```python
right_limit=Q3+2*IQR
right_limit
```

```python

cond=((df["price"])<(left_limit))| ((df["price"])>(right_limit))
df[cond]
Outliers=df[cond]
Outliers
```

```python
Average_IQR = df[~cond]
```

```python
Average_IQR
```

```python
IQR_Sampled=Average_IQR.sample(frac=0.05,replace=True,random_state=1)
IQR_Sampled
```
